# Gujarat Election 2022 Result Analysis

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('gujart_election.csv')
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
#convert first character of each word to upper case
dataset['Candidate'] = dataset['Candidate'].apply(lambda x: x.title())
```

```python
dataset.head(15)
```

```python
dataset['Rank'] = dataset.groupby('Vidhan_Sabha_Name')['Total Votes'].rank(method='min',ascending= False).sort_values()
```

```python
dataset.head(20)
```

## Total number of seats win by each party

```python
dataset[dataset['Rank']==1]['Party'].value_counts()
```

```python
dataset.sort_values(['Vidhan_Sabha_Name','Rank'],inplace=True)
```

```python
#dataset.reset_index(inplace=True)
```

```python
dataset.to_csv('Edited_gujart_election_data.csv',index=False)
```

```python
dataset.head(20)
```

```python
dataset['margin'] = dataset['Total Votes'].apply(lambda x: x-x)
```

```python
dataset
```

```python
dataset['margin'] = dataset['Total Votes'] - dataset['Total Votes'].shift(-1)
```

### Create Winning Candidate Dataset

```python
dataset_winning = dataset[dataset['Rank']==1]
```

```python
dataset_winning.reset_index(inplace=True)
```

```python
dataset_winning.info()
```

```python
dataset_winning = dataset_winning.drop(['level_0','index'],axis=1)
```

```python
dataset_winning.sort_values('margin',inplace=True)
```

```python
dataset_winning['Vidhan_Sabha'] = dataset_winning['Vidhan_Sabha_Name'].str.split('-',expand=True)[1]
```

```python
dataset_winning.head()
```

```python
dataset_winning.to_csv('dataset_winning.csv',index=False)
```

## Top 10 candidate who win with low margin

```python
top_10_low_margin_winning = dataset_winning.head(10)
```

## Top 10 candidate who win with high margin

```python
top_10_high_margin_winning = dataset_winning.sort_values('margin',ascending=False).head(10)
```
